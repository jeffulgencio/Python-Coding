```python
import pandas as pd
```

```python
df = pd.read_csv('AirBnB_Bathroom_Bedroom.csv')
```

```python
df
```

```python
df.groupby(['city','property_type']).mean().reset_index().rename(columns=\
{'bedrooms':'n_bedrooms_avg','bathrooms':'n_bathrooms_avg'}).drop(columns=['id','price','accommodates','cleaning_fee',\
'number_of_reviews','review_scores_rating','zipcode','beds'])
```

```python
playbook_events = pd.read_csv('Apple_EventName.csv')
```

```python
playbook_events
```

```python
pd.DataFrame(playbook_events.loc[playbook_events.device.str.upper() == 'MACBOOK PRO',['event_name','occurred_at']].groupby('event_name')\
['occurred_at'].count()).reset_index().sort_values(by='occurred_at',ascending=False).rename(columns={'occurred_at':'event_count'})
```
